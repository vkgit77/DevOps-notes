###### 1. Git Checkout:- Discarding local changes (before staging) 

```python
It happens that you modify a file in your local working directory and
sometimes wish just to discard the committed changes. 

Here is when the checkout command will help you.
```

```python
Create a new file with name 'HELLO.html'.
Insert the code as given below and save it to staging/indexing. (git add.)
```

```python
FILE: HELLO.html
------------------    
<html>
  <head>
  </head>
  <body>
    <h1>Hello, World!</h1>
  </body>
</html>
```

```python
Make changes to the hello.html file in the form of an unwanted comment.

FILE: HELLO.html
------------------    
<html>
  <head>
  </head>
  <body>
    <h1>Hello, World!</h1>
    <!-- This is a bad comment.  We want to revert it. -->
  </body>
</html>
```

```python
Check the status:-
-------------------
First, check the working directory’s status.

$ git status
```

```python
We see that the hello.html file has been modified, but not staged yet.

Undoing the changes in the working directory:-
-----------------------------------------------

Use the checkout command in order to checkout the repository’s version of the hello.html file.

$ git checkout hello.html

$ git status

$ cat hello.html
```

```python
The status command shows there were no unstaged changes in the working directory. 
And the “bad comment” is no longer contained in the file.
```

###### 2. Git Reset :- Cancel Staged changes (before committing)

```python
Edit file and stage changes:-
-----------------------------
Make changes to the hello.html file in the form of an unwanted comment

FILE: HELLO.HTML
----------------
<html>
  <head>
    <!-- This is an unwanted but staged comment -->
  </head>
  <body>
    <h1>Hello, World!</h1>
  </body>
</html>

Stage the modified file.
------------------------
$ git add hello.html

Check the status

$ git status

Status shows that the change has been staged and is ready to commit.
```

```python
Reset the buffer zone:-
-----------------------
Fortunately, the displayed status shows us exactly what we should do to cancel staged changes.

$ git reset HEAD hello.html
```

```python
The reset command resets the buffer zone to HEAD. 
This clears the buffer zone from the changes that we have just staged.

The reset command (default) does not change the working directory.
Therefore, the working directory still contains unwanted comments.
We can use the checkout command to remove unwanted changes from working directory.
```

```python
Switch to commit version
--------------------------
$ git checkout hello.html

$ git status

# On branch master
nothing to commit (working directory clean)
Our working directory is clean again.
```

###### 3. Git Reset

```python
The term reset stands for undoing changes.
The git reset command is used to reset the changes. 
The git reset command has three core forms of invocation.

These forms are as follows.

- Mixed
- Soft
- Hard
```

###### git reset mixed

```python
git reset mixed -- commit id  (This is default option of reset)
-----------------------------
 - It will move the files to working directory, but wont change the contents. 
   Also, it will remove the commit id from the log. 

Steps:-
--------
- Create a new file and add the above given html code. 
- Now add it to staging and commit it. 
- Now git log. it will show one commit. 
- Now open that file and add a new line to the html code. ( <h1>Amazon products</h1>)
- Add it to staging and commit it. 
- Now git log. it will show two commits. 
- Now reset to previous commit id. $ git reset --mixed commit id. 
- Now check the git log. we can see only one commit and file will be moved to staging to working directory
  and contents in the file will not be changed. 
---------------------------------------------------

Steps:-
-------
▪ Modify contents in two files and add & commit them. 
▪ Now check the git log. Ensure we have more than one commit in log. 
▪ Now reset to previous commit id. $ git reset --mixed commit id
▪ Once done, check git status -  it will move the files to working directory. 
▪ Now check the git log. We can see only one commit.
```

###### git reset soft 

```python
git reset soft -- commit id
----------------------------
 - It will move the files to staging, but wont change the contents. 
   Also, it will remove the commit id from the log. 

Steps:-
--------
- Create a new file and add the above given html code. 
- Now add it to staging and commit it. 
- Now git log. it will show one commit. 
- Now open that file and add a new line to the html code. ( <h1>Amazon products</h1>)
- Add it to staging and commit it. 
- Now git log. it will show two commits. 
- Now reset to previous commit id. $ git reset -- soft commit id. 
- Now check the git log. we can see only one commit and file will be moved to staging
  and contents in the file will not be changed. 
----------------------------------------------------------------------

▪ Modify contents in two files and add & commit them. 
▪ Now check the git log. Ensure we have more than one commit in log. 
▪ Now reset to previous commit id. $ git reset --soft commit id
▪ Once done, check git status -  it will move the files to Staging. 
▪ Now check the git log. We can see only one commit.
```

###### git reset hard

```python
git reset hard commit id 
-------------------------

 - It will remove the changes made in the latest commit ID and 
   reset the files to the commit id what we selected. 
   Also, it will remove the commit id from the log. 

Steps:-
--------
▪ Modify contents in two files and add & commit them. 
▪ Now check the git log. Ensure we have more than one commit in log. 
▪ Now reset to previous commit id. $ git reset --hard commit id
▪ Once done, check git status -  it will remove the files from Staging & working directory and also the changes made. 
▪ Now check the git log. We can see only one commit.

Git reset hard is a command which should be used carefully as
it can delete your entire code commit history and what you have done till a particular commit. 
Therefore it should be used with certain caution.

This command can be dangerous to use. Only use if you know what exactly you are doing. 
```

###### 3. Git Revert

```python
Sometimes you realize that the new commits are wrong, and you want to cancel them.
To cancel the commit we will create a new commit, cancelling the unwanted changes.

Edit the file and make a commit:-
---------------------------------
Replace hello.html with the following file.

FILE: HELLO.HTML
-----------------
<html>
  <head>
  </head>
  <body>
    <h1>Hello, World!</h1>
    <!-- This is an unwanted but committed change -->
  </body>
</html>

$ git add hello.html

$ git commit -m "we didn't want this commit"

Make a commit with new changes that discard previous changes:-
----------------------------------------------------------------
To cancel the commit, we need to create a commit that deletes the changes saved by unwanted commit.

$ git revert HEAD

Go to the editor, where we can edit the default commit message or leave it as is.
Save and close the file.

we will see …

RESULT:
--------
$ git revert HEAD

[master 45fa96b] Revert "we didn't want this commit"

1 files changed, 1 insertions(+), 1 deletions(-)
    
Since we have cancelled the last commit, we can use HEAD as the argument for cancelling. 
We may cancel any random commit in history, pointing out its hash value.
```

###### 4. Git Amend:- Changing commits

```python
Put an author comment on the page. Adding first line :- Author: Alexander Shvets

FILE: HELLO.HTML
-----------------
<!-- Author: Alexander Shvets -->
<html>
  <head>
  </head>
  <body>
    <h1>Hello, World!</h1>
  </body>
</html>

$ git add hello.html

$ git commit -m "Add an author comment"

Oops... email required
-----------------------
After making the commit, you understand that every good comment
should include the author's email. 
Edit the hello page to provide an email. Adding email (alex@githowto.com) in the first line. 

FILE: HELLO.HTML
-----------------
<!-- Author: Alexander Shvets (alex@githowto.com) -->
<html>
  <head>
  </head>
  <body>
    <h1>Hello, World!</h1>
  </body>
</html>

Change the previous commit:
----------------------------
We do not want to create another commit for adding the e-mail address.
Let us change the previous commit and add an e-mail address.

$ git add hello.html

$ git commit --amend -m "Add an author/email comment"
```

```python
$ git log --oneline

The new "author/email" commit replaces the original "author" commit. 
The same effect can be achieved by resetting the last commit in the branch, and recommitting new changes.
```

###### detach commit

```python
If you want to temporarily go back to it, fool around, 
then come back to where you are, all you have to do is check out the desired commit:
```

```python
HEAD is simply a reference to the current commit (latest) on the current branch. 
There can only be a single HEAD at any given time (excluding git worktree).

The content of HEAD is stored inside .git/HEAD, 
and it contains the 40-bytes SHA-1 hash of the current commit.
```

###### detached HEAD

```python
If you are not on the latest commit - meaning that HEAD is pointing 
to a prior commit in history it's called detached HEAD.
```

```python
On the command-line it will look like this - 
SHA-1 hash instead of the branch name since the 
HEAD is not pointing to the the tip of the current branch:
```

```python
A few options on how to recover from a detached HEAD:
    
git checkout
git checkout <commit_id>
git checkout -b <new branch> <commit_id>
git checkout HEAD~X // x is the number of commits t go back

This will checkout new branch pointing to the desired commit. This command will checkout to a given commit.

At this point you can create a branch and start to work from this point on:

# Checkout a given commit.
# Doing so will result in a `detached HEAD` which mean that the `HEAD`
# is not pointing to the latest so you will need to checkout branch
# in order to be able to update the code.

git checkout <commit-id>

# Create a new branch forked to the given commit

git checkout -b <branch name>
```

```python
git reflog
----------

You can always use the reflog as well. 
git reflog  will display any change which updated the HEAD and 
checking out the desired reflog entry will set the HEAD back to this commit.

Every time the HEAD is modified there will be a new entry in the reflog

git reflog

git checkout HEAD@{...}

This will get you back to your desired commit.
```

###### git Stash

```python
Sometimes you want to switch the branches,
but you are working on an incomplete part of your current project. 
You don't want to make a commit of half-done work. 
Git stashing allows you to do so. 
The git stash command enables you to switch branches without committing the current branch.
```

```python
Generally, the stash's meaning is "store something safely in a hidden place."
The sense in Git is also the same for stash; Git temporarily saves your data safely without committing.
```
